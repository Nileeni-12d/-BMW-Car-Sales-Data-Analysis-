**Analysis and Visualizations**

**1. Pricing Analysis**


```python
avg_price_fuel = df.groupby('Fuel_Type', dropna=False)['Price_USD'].mean().reset_index().sort_values('Price_USD', ascending=False)

plt.figure(figsize=(7,4))
sns.barplot(data=avg_price_fuel, x='Fuel_Type', y='Price_USD', order=avg_price_fuel['Fuel_Type'],color='Lightblue')
plt.title('Average BMW Price by Fuel Type')
plt.ylabel('Average Price (USD)')
plt.xlabel('')
plt.tight_layout()
plt.savefig('Price Analysis.jpg')
plt.show()


avg_price_fuel 
```

📌 Insight:
Electric BMWs have the highest average price at $75,276, followed closely by Diesel ($75,080) and Petrol ($74,990). Hybrids have the lowest average price at $74,798.

📌 Extended Insight:
The price gap between the most expensive (Electric) and least expensive (Hybrid) fuel types is minimal — less than $500. This suggests BMW maintains a relatively consistent pricing strategy across fuel types, possibly to avoid large perception gaps in value.

📌 Possible Reason:

BMW’s positioning of electric models as premium, tech-forward vehicles with higher R&D costs.

Diesel and petrol models priced closely due to similar performance and luxury positioning.

Hybrids possibly priced slightly lower to attract eco-conscious buyers into BMW’s portfolio without premium deterrents.

📌 Business Implication:

Opportunity to command a stronger price premium on electric models if unique features (autonomous driving, extended range, luxury interiors) are highlighted more.

Maintain competitive pricing for hybrids to capture growing sustainability-focused segments.

Since pricing variation is minimal, differentiation should come from features, performance, and ownership benefits rather than just cost.

📌 Insight:
Price distributions vary noticeably by Fuel Type, with Electric and Diesel showing greater price variability and more outliers compared to Petrol and Hybrid.

📌 Extended Insight:
The wider spread and outliers in Electric and Diesel categories indicate a mix of premium and standard models, whereas Petrol and Hybrid prices cluster more tightly, reflecting more consistent pricing.

📌 Possible Reason:
Outliers may result from limited editions, special features, or exceptionally maintained vehicles. Electric vehicles often come with advanced technology, and Diesel models span from economy to luxury, causing price variation.

📌 Business Implication:
Identifying these outliers and price ranges enables better pricing strategies, inventory management, and targeted marketing by highlighting premium options and preventing mispricing.



**6.  Vehicle Price Variation by Mileage Efficiency**

```python
# Assuming 'Mileage_KM' is total km driven and 'Price_USD' is price
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Mileage_KM', y='Price_USD', hue='Fuel_Type', alpha=0.7)
plt.title('Mileage Efficiency vs Price by Fuel Type')
plt.xlabel('Mileage (km)')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.savefig("Mileage Efficiency vs Price.jpg")
plt.show()
```

```python
# Summary statistics of Price vs Mileage by Fuel_Type
summary = df.groupby('Fuel_Type').agg({
    'Mileage_KM': ['mean', 'median', 'min', 'max'],
    'Price_USD': ['mean', 'median', 'min', 'max']
})
summary.round(2)
```

📌 Insight:
There is a general negative correlation between mileage and price, indicating that vehicles with higher mileage tend to have lower prices.

📌 Extended Insight:
Electric and Hybrid vehicles show relatively higher prices at comparable mileage levels compared to Diesel and Petrol, suggesting better mileage efficiency or perceived value retention.

📌 Possible Reason:
Electric and Hybrid cars may have newer technology and better efficiency, which supports higher prices despite mileage. Diesel and Petrol vehicles depreciate faster with mileage.

📌 Business Implication:
Dealerships can leverage mileage efficiency as a selling point for Electric and Hybrid models, emphasizing their long-term value and cost savings to customers.
